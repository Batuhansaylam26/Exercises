```python
import pandas as pd
import numpy as np
data=pd.read_csv("C:\\Users\\ASUS\\Downloads\\amazon_sales_dataset.csv")
data=data[data.Store==1]
data=data[ data.Dept==1]
data=data.drop(["max","min","mean","median","std","Unnamed: 3","Type"],axis=1)
data=data.reset_index()
Q1 = np.percentile(data.Weekly_Sales, 25,
                interpolation = 'midpoint')
 
Q3 = np.percentile(data.Weekly_Sales, 75,
                interpolation = 'midpoint')
IQR = Q3 - Q1
    
 


# Upper bound
upper = np.where(data.Weekly_Sales >= (Q3+1.5*IQR))
# Lower bound
lower = np.where(data.Weekly_Sales <= (Q1-1.5*IQR))
 
''' Removing the Outliers '''
data.drop(upper[0], inplace = True)
data.drop(lower[0], inplace = True)
data=data.reset_index()
data["y"]=data.Weekly_Sales
data['Time'] = np.arange(len(data.Date))
#data.index=data.ds
data.head()
```

```python
data["ds"] = pd.to_datetime(data.Date,format='%d/%m/%Y')
data["ds"]=data["ds"].dt.strftime('%d/%m/%Y')
```

```python
df=data[["ds","y"]]
df.index=df["ds"]
df
```

```python
from statsmodels.tsa.stattools import adfuller
from numpy import log
result = adfuller(np.log(df.y).dropna())
print('ADF Statistic: %f' % result[0])
print('p-value: %f' % result[1])
```

```python
import numpy as np, pandas as pd
from statsmodels.graphics.tsaplots import plot_acf, plot_pacf
import matplotlib.pyplot as plt

# Original Series
fig, axes = plt.subplots(4, 2, sharex=True)
axes[0, 0].plot(df.ds,df.y); axes[0, 0].set_title('Original Series')
plot_acf(df.y, ax=axes[0, 1])

# 1st Differencing
axes[1, 0].plot(df.ds,df.y.diff()); axes[1, 0].set_title('1st Order Differencing')
plot_acf(df.y.diff().dropna(), ax=axes[1, 1])

# 2nd Differencing
axes[2, 0].plot(df.ds,df.y.diff().diff()); axes[2, 0].set_title('2nd Order Differencing')
plot_acf(df.y.diff().diff().dropna(), ax=axes[2, 1])

axes[3, 0].plot(df.ds,np.log(df.y)); axes[2, 0].set_title('2nd Order Differencing')
plot_acf(np.log(df.y).dropna(), ax=axes[3, 1])
plt.show()
```

```python
# PACF plot of 1st differenced series
plt.rcParams.update({'figure.figsize':(9,3), 'figure.dpi':120})

fig, axes = plt.subplots(1, 2, sharex=True)
axes[0].plot(np.log(df.y)); axes[0].set_title('y ')
axes[1].set(ylim=(0,5))
plot_pacf(np.log(df.y).dropna(), ax=axes[1])

plt.show()
```

```python
import statsmodels.api as sm
```

```python
model = sm.tsa.arima.ARIMA(df.y, order=(1,1,2))
model_fit = model.fit()
print(model_fit.summary())
```

```python
import matplotlib.pyplot as plt
residuals = model_fit.resid[1:]
fig, ax = plt.subplots(1,2)
residuals.plot(title='Residuals', ax=ax[0])
residuals.plot(title='Density', kind='kde', ax=ax[1])
plt.show()
```

```python
from statsmodels.graphics.tsaplots import plot_predict


plot_predict(model_fit,dynamic=False)
plt.show()
```
