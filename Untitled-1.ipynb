```python
m= NeuralProphet()
```

```python
Date=m.fit(df,freq="W",epochs=1000)
```

```python
df_train, df_test = m.split_df(df, freq='D')
metrics = m.fit(df_train, freq='D', epochs=1000, validation_df=df_test)
```
